```python
#initial dependencies  
import pandas as pd
import numpy as np
```

```python
#initialise the dataset
restaurants = pd.read_csv("BangaloreZomatoData.csv")
restaurants.head()
```

```python
#we have 8923 restaurants
restaurants.shape
```

```python
#listing out the columns we require in our project
#Name
#Cuisines
#PopularDishes

restaurants = restaurants[['Name','Cuisines',"PopularDishes"]]
restaurants.head()
```

```python
#Now we need to group all of these features into a single tag columns as strings
```

```python
#counting the NaN values in popular dishes
restaurants["PopularDishes"].isna().sum()
# lets drop them
restaurants = restaurants.dropna(subset=["PopularDishes"])
restaurants = restaurants.reset_index(drop=True)
```

```python
restaurants["PopularDishes"].isna().sum()
```

```python
restaurants
```

```python
#but now there may be multiple restaurants with same name lets count them
restaurants.duplicated("Name").sum() #568 of them, we have to drop
restaurants = restaurants.drop_duplicates(subset="Name")
restaurants = restaurants.reset_index(drop=True)
```

```python
restaurants.duplicated("Name").sum()
```

```python
#converting the ratings and cost to strings
# restaurants["AverageCost"] =  restaurants["AverageCost"].astype(str)
# restaurants["Delivery Ratings"] =  restaurants["Delivery Ratings"].astype(str)
restaurants["Cuisines"] =  restaurants["Cuisines"].astype(str)
restaurants["PopularDishes"] =  restaurants["PopularDishes"].astype(str)
# restaurants["rest_type"] =  restaurants["rest_type"].astype(str)

# restaurants["dish_liked"] =  restaurants["dish_liked"].astype(str)
restaurants
```

```python
#turning into list
restaurants["Cuisines"]  = restaurants["Cuisines"].apply(lambda x:x.split(", "))
# restaurants["Delivery Ratings"]  = restaurants["Delivery Ratings"].apply(lambda x:x.split(", "))
# restaurants["KnownFor"]  = restaurants["KnownFor"].apply(lambda x:x.split(", "))
restaurants["PopularDishes"]  = restaurants["PopularDishes"].apply(lambda x:x.split(", "))
# restaurants["PeopleKnownFor"]  = restaurants["PeopleKnownFor"].apply(lambda x:x.split(", "))
# restaurants["rest_type"]  = restaurants["rest_type"].apply(lambda x:x.split(", "))
# restaurants["dish_liked"]  = restaurants["dish_liked"].apply(lambda x:x.split(", "))
```

```python
restaurants
```

```python
#now i want to remove all the unnecessary spaces between the values which will create confusion inour model
restaurants["Cuisines"] =  restaurants["Cuisines"].apply(lambda x : [i.replace(" ","") for i in x])
# restaurants["Delivery Ratings"] =  restaurants["Delivery Ratings"].apply(lambda x : [i.replace(" ","") for i in x])
# restaurants["KnownFor"] =  restaurants["KnownFor"].apply(lambda x : [i.replace(" ","") for i in x])
restaurants["PopularDishes"] =  restaurants["PopularDishes"].apply(lambda x : [i.replace(" ","") for i in x])
# restaurants["PeopleKnownFor"] =  restaurants["PeopleKnownFor"].apply(lambda x : [i.replace(" ","") for i in x])
# restaurants["rest_type"] =  restaurants["rest_type"].apply(lambda x : [i.replace(" ","") for i in x])
# restaurants["dish_liked"] =  restaurants["dish_liked"].apply(lambda x : [i.replace(" ","") for i in x])
restaurants
```

```python
#now lets create the tags columns which will have all the columns
restaurants["tags"] = restaurants["Cuisines"] + restaurants["PopularDishes"]
restaurants["tags"]
```

```python
#hey there are some empty strings which can be trouble some lets remove them
restaurants['tags'] = restaurants['tags'].apply(lambda x: list(filter(None, x)))
#removing all the "-", this was for those restaurants which dont have a rating
# restaurants['tags'] = [list(filter(lambda x: x != '-', lst)) for lst in restaurants['tags']]
restaurants['tags']
```

```python
new_df = restaurants[['Name','tags']]
new_df
```

```python
#convert list to strings 
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#lower case (recommended)
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags']
```

```python
#now for the main stuff, we convert the strings into vectors for our recommendation
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
array =  cv.get_feature_names_out()
for i in array:
    print(i)
```

```python
#recommender function
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
#lets find out for 1st restaurant
sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(restaurant):
    index = new_df[new_df['Name'] == restaurant].index[0]
    distances = similarity[index]
    ourlist = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]

    for i in ourlist:
        print(new_df.iloc[i[0]].Name)
        # print(i[0])
```
